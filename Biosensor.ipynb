## Simulating lead sensitive gene modelled by the 2015 iGEM Bielefeld Team

[Bielefeld CeBiTec Biosensor Model](http://2015.igem.org/Team:Bielefeld-CeBiTec/Modeling/Biosensor)

$\textbf{Reactions in the system}$ 

$G_R \xrightarrow{} m_R$ ...................................(1)

$m_R \xrightarrow{k_{TL2}} R$ .................................(2)

$2R \underset{k_{-2R}}{\stackrel{k_{2R}}{\rightleftharpoons}} R_2$ ..................................(3)

$R_2+O \underset{k_-r}{\stackrel{k_r}{\rightleftharpoons}} {R_2}O$ .........................(4)

$2A + R_2 \underset{k_{-dr1}}{\stackrel{k_{dr1}}{\rightleftharpoons}} {R_2}{A_2}$ ...................(5)

$2A +{R_2}O \underset{k_{-dr2}}{\stackrel{k_{dr2}}{\rightleftharpoons}} {R_2}{A_2}+O$ ........(6)

$O \xrightarrow{} O+ m_F$ ..............................(7)

${R_2}O \xrightarrow{k_{leak}} {R_2}O + m_F$ ................(8)

$m_F \xrightarrow{k_{TL1}} F_{in}$ ...............................(9)

$F_{in} \underset{-}{\stackrel{k_{mat}}{\rightleftharpoons}} F$ .....................................(10)

$m_F \xrightarrow{\lambda_{m_1}} \phi$ ...................................(11)

$m_R \xrightarrow{\lambda_{m_2}} \phi$ ...................................(12)

| Symbol | Species | Initial Value (nM) |
| ----- | ----- | ----- |
| $G_R$ | Repressor Gene | 8 |
| $m_R$ | Repressor mRNA | 0 |
| $R$ | Repressor Monomer | 0 |
| $R_2$ | Repressor Dimer | 0 |
| $O$ | Free Operator of Reporter Gene | 8 |
| ${R_2}O$ | Repressor-Operator Complex | 0 |
| $A$ | Analyte (Lead) | $0$ to $10^5$ |
| ${R_2}{A_2}$ | Inactive Repressor | 0 |
| $m_F$ | Reporter mRNA | 0 |
| $F_{in}$ | Inactive Reporter | 0 |
| $F$ | Reporter (GFP) | 0 |
| $TLR$ | Translation Resources | $1520$ |

$T_O=O+{R_2}O$ - Total Operator Concentration 

$T_A=A+A_{out}+2{R_2}{A_2}$ - Total Analyte Concentration

----
$\textbf{Equations}$

$\frac{d[m_R]}{dt} = v_{TX2}\frac{[G_R]^2}{K_{TX2}^2 + [G_R]^2}-{\lambda_{m_R}}[m_R]$ <br/>where $m_R$ is the repressor mRNA, $v_{TX2}$ is the max transcription (1) rate, $\lambda_{m_R}$ is the degradation constant<br/><br/>

$\frac{d[R]}{dt} = k_{TL2}[TLR]\frac{[m_R]}{{K_{TL2}}+{[m_R]}} - 2k_{2R}[R]^2 + 2k_{-2R}[R_2]$<br/>where $R$ is the repressor protein, $k_{TL2}$ is the translation rate, $k_{2R}$ and $k_{-2R}$ are repressor protein dimer formation (3) constants<br/><br/>

$\frac{d[R_2]}{dt} = k_{2R}[R]^2 - k_{-2R}[R_2] - k_r{[R_2]}[O] + k_{-r}(T_O - [O]) - k_{dr1}[A]^2{[R_2]} + k_{-dr1}\frac{(T_A - [A] - [A_{out}])}{2}$<br/>where $R_2$ is the repressor protein dimer, $k_r$ and $k_{-r}$ are repression mechanism (4) constants, $k_{dr1}$ and $k_{-dr1}$ are the first derepression mechanism (5) constants<br/><br/>

$\frac{d[A_{out}]}{dt} = -P*([{A_{out}]-[A]})$

$\frac{d[A]}{dt} = P*({[A_{out}]-[A]}) - 2k_{dr1}[A]^2{[R_2]} + 2k_{-dr1}\frac{(T_A - [A] - [A_{out}])}{2} -2k_{dr2}[A]^2(T_O - [O]) + 2k_{-dr2}[O]\frac{(T_A - [A] - [A_{out}])}{2}$<br/>where $A$ is the analyte (lead), $k_{dr2}$ and $k_{-dr2}$ are the second derepression mechanism (6) constants<br/><br/>

$\frac{d[O]}{dt} = - k_r{[R_2]}[O] + k_{-r}(T_O - [O]) -k_{dr2}[A]^2(T_O - [O]) + k_{-dr2}[O]\frac{(T_A - [A] - [A_{out}])}{2}$<br/>where $O$ is the free operator concentration<br/><br/>

$\frac{d[m_F]}{dt} = v_{TX1}\frac{[O]^2}{K_{TX1}^2 + [O]^2}-{\lambda_{m_F}}[m_R] + k_{leak}(T_O - [O])$<br/>where $m_F$ is the reporter mRNA, $v_{TX1}$ is the max transcription (7) rate, $\lambda_{m_F}$ is the degradation constant, $k_{leak}$ is the leaky transcription (8) rate<br/><br/>

$\frac{d[F_{in}]}{dt} = k_{TL1}[TLR]\frac{[m_F]}{{K_{TL1}}+{[m_F]}} - k_{mat}[F_{in}]$<br/>where $F_{in}$ is the inactive reporter protein (GFP), $k_{TL1}$ is the translation rate, $k_{mat}$ is the GFP maturation rate (10)<br/><br/>

$\frac{d[F]}{dt} = k_{mat}[F_{in}]$<br/>where $F$ is the activated GFP concentration<br/><br/>

$\frac{d[TLR]}{dt} = -v_{\lambda TLR}\frac{[TLR]}{{K_{\lambda TLR}}+[TLR]}$<br/>where $TLR$ is the concentration of translation resources, $v_{\lambda TLR}$ is degradation rate of the translation resources<br/><br/>

## Lead Biosensor Time Series 

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pylab as plot
from decimal import Decimal
from scipy.integrate import solve_ivp
import math as mt
from matplotlib.pyplot import figure

def safe_division(x, y):
    if y == 0:
        return 0
    else:
        return x / y

#Initialize some constants
vTX1 = 18.2       #reporter transcription rate constant
vTX2 = 18.2       #repressor transcription rate constant
KTX1 = 8.5        #Michaelis-Menten constant for reporter transcription 
KTX2 = 8.5        #Michaelis-Menten constant for repressor transcription
lam_m1 = 0.08     #reporter mRNA degradation rate constant
lam_m2 = 0.08     #repressor mRNA degradation rate constant
#vTL = 16.1       #Reporter translation rate constant, 0.0076
KTL1 = 29.9       #Michaelis-Menten constant for translation of reporter
KTL2 = 29.9       #Michaelis-Menten constant for translation of repressor
vlamTLR = 13.5    #Translation resources degradation rate constant
KlamTLR = 53.2    #Michaelis-Menten constant for degradation of TL resources
kmat = 0.1155      #Reporter maturation rate constant, sfGFP = 1/6 min
kgd = 0.001155       #GFP Degradation rate
k2R = 50          #repressor dimerization rate constant
k_2R  = 0.001     #repressor dimer dissociation rate constant
kr = 960          #association rate constant for repression
k_r = 2.4         #dissociation rate constant for repression
kdr1 = 3.0e7      #association rate constant for first derepression mechanism 
k_dr1 = 12        #dissociation rate constant for first derepression mechanism
kdr2 = 3.0e7      #association rate constant for second derepression mechanism   
k_dr2 = 4800      #dissociation rate constant for second derepression mechanism
kleak = 0.0033    #leak reporter transcription rate constant
P = 400*1e-4      #Permeability, for now using P of Na+ ~ 7.4*1e-4 nm/s

def model(z,t):
    
    #Initialize variables in a vector
    mR = z[0]         #repressor mRNA
    R = z[1]          #repressor monomer
    R2 = z[2]         #repressor dimer               #R2O - free operator of reporter gene
    Aout = z[3]
    A = z[4]          #analyte                       #R2A2 - repressor-operator complex
    O = z[5]          #inactive repressor
    mF = z[6]         #reporter mRNA
    Fin = z[7]        #inactive reporter
    F = z[8]          #reporter
    TLR = z[9]        #Translation resources
    TA = z[10]
    vTL = z[11]
    
    
    #ODEs  
    dmRdt = (vTX2*(GR**2)/((KTX2**2)+(GR**2))) - lam_m2*mR                                       #Repressor mRNA Conc
    dRdt = (vTL*TLR*(mR**3)/((KTL2**3)+(mR**3)+(mF**3))) - 2*k2R*(R**2) + 2*k_2R*R2             #Repressor Conc  
    dR2dt = k2R*(R**2) - k_2R*R2 - kr*R2*O + k_r*(TO-O) - kdr1*(A**2)*R2 + k_dr1*((TA-A-Aout)/2)      #Repressor dimer Conc               
    dAoutdt = -P*(Aout-A)
    dAdt = P*(Aout-A) - 2*kdr1*(A**2)*R2 + 2*k_dr1*((TA-A-Aout)/2) - 2*kdr2*(A**2)*(TO-O) + 2*k_dr2*O*((TA-A-Aout)/2)  #Analyte Conc
    dOdt = -kr*R2*O + k_r*(TO-O) + kdr2*(A**2)*(TO-O) - k_dr2*O*((TA-A-Aout)/2)                       #Operator Conc
    dmFdt = (vTX1*(O**2)/((KTX1**2)+(O**2))) - lam_m1*mF + kleak*(TO-O)                          #Reporter mRNA Conc
    dFindt = (vTL*TLR*(mF**3)/((KTL1**3)+(mF**3)+(mR**3))) - kmat*Fin                           #Inactive Reporter Conc
    dFdt = kmat*Fin - kgd*F                                                                             #Active Reporter Conc  
    dTLRdt = -vlamTLR*TLR/((KlamTLR)+(TLR))                                                      #Translation Resources Conc
    dTAdt = 0
    dvTLdt = 0
    
    return [dmRdt, dRdt, dR2dt, dAoutdt, dAdt, dOdt, dmFdt, dFindt, dFdt, dTLRdt, dTAdt, dvTLdt]


#Initial Conditions, conc in nM
mR0 = 0
R0 = 0
R20 = 0
R21 = 300
R2O0 = 0
R2A20 = 0
mF0 = 0
Fin0 = 0
F0 = 0
TLR0 = 1520

O0 = 5
GR = O0

A0 = 0                 #Lead Concentration
Aout0 = 7200 

TO = O0 + R2O0        #LOOK AT THIS----------------------<<<<<<
TA = A0 + Aout0 + 2*R2A20

v = 0.082

t = np.linspace(0,250,100000)

z = [mR0,R0,R20,5000,A0,O0,mF0,Fin0,F0,TLR0,TA,v]
z1 = [mR0,R0,R20,500,A0,O0,mF0,Fin0,F0,TLR0,TA,v]
z2 = [mR0,R0,R20,50,A0,O0,mF0,Fin0,F0,TLR0,TA,v]

ans = odeint(model,z,t)#, t_eval = np.linspace(0,300,100000))
ans1 = odeint(model,z1,t)
ans2 = odeint(model,z2,t)

a = ans[:,8]
g = []
g[:] = [A*0.6022 for A in a]

a1 = ans1[:,8]
g1 = []
g1[:] = [A*0.6022 for A in a1]

a2 = ans2[:,8]
g2 = []
g2[:] = [A*0.6022 for A in a2]

#plt.ylim(0,2000)
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.ylabel("GFP (# of molecules)")
plt.xlabel("Time(min)")
#plt.title("Lead = 7.2 μM, LSODA")
plt.plot(t, g, 'b-', label = 'Lead = 5μM')
plt.plot(t, g1, 'r-', label = 'Lead = 0.5μM')
plt.plot(t, g2, 'g-', label = 'Lead = 0.05μM')
plt.legend()
plt.show()

print(np.amax(g))
```
