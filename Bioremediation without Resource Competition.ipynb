$\textbf{Reactions in the system}$ 

$G_R \xrightarrow{} m_R$ ...................................(1)

$m_R \xrightarrow{k_{TL2}} R$ .................................(2)

$2R \underset{k_{-2R}}{\stackrel{k_{2R}}{\rightleftharpoons}} R_2$ ..................................(3)

$R_2+O \underset{k_-r}{\stackrel{k_r}{\rightleftharpoons}} {R_2}O$ .........................(4)

$2A + R_2 \underset{k_{-dr1}}{\stackrel{k_{dr1}}{\rightleftharpoons}} {R_2}{A_2}$ ...................(5)

$2A +{R_2}O \underset{k_{-dr2}}{\stackrel{k_{dr2}}{\rightleftharpoons}} {R_2}{A_2}+O$ ........(6)

$O \xrightarrow{} O+ m_F$ ..............................(7)

${R_2}O \xrightarrow{k_{leak}} {R_2}O + m_F$ ................(8)

$m_F \xrightarrow{k_{TL1}} F_{in}$ ...............................(9)

$F_{in} \underset{-}{\stackrel{k_{mat}}{\rightleftharpoons}} F$ .....................................(10)

$m_F \xrightarrow{\lambda_{m_1}} \phi$ ...................................(11)

$m_R \xrightarrow{\lambda_{m_2}} \phi$ ...................................(12)

$m_T \xrightarrow{k_{TL4}} T$

$A_{out} \underset{k_{-in}}{\stackrel{T*k_{in}}{\rightleftharpoons}} A$

$m_D \xrightarrow{k_{TL3}} D$

$D + nA \underset{k_{-seq}}{\stackrel{k_{seq}}{\rightleftharpoons}} DA_n$

$m_D \xrightarrow{\lambda_{m_3}} \phi$

$m_T \xrightarrow{\lambda_{m_4}} \phi$

$\textbf{Equations}$ 

$\frac{d[m_R]}{dt} = v_{TX2}\frac{[G_R]^2}{K_{TX2}^2 + [G_R]^2}-{\lambda_{m_R}}[m_R]$<br/>
	<br/>
Protein translation represented by Michaelis-Menten<br/>
$\frac{d[R]}{dt} = k_{TL2}[TLR]\frac{[m_R]}{{K_{TL2}}+{[m_R]}} - 2k_{2R}[R]^2 + 2k_{-2R}[R_2]$<br/><br/>
	In the equation below $\frac{(T_A-[A_{out}]-[A]-n[DA_n])}{2}$ is a substitute for ${R_2}{A_2}$.<br/><br/>
	$\frac{d[R_2]}{dt} = k_{2R}[R]^2 - k_{-2R}[R_2] - k_r{[R_2]}[O] + k_{-r}(T_O - [O]) - k_{dr1}[A]^2{[R_2]} + k_{-dr1}\frac{(T_A-[A_{out}]-[A]-n[DA_n])}{2}$<br/><br/>
$T_{mol}$ is the number of protein molecules of pbrT. It can be calculated by multiplying the concentration of pbrT with Avagadro number and dividing by cell volume. Cell volume is assumed to be 1fL $(10^{-15}L$.<br/>
	$\frac{d[A_{out}]}{dt} = -P([{A_{out}]-[A]}) - T_{mol}v_{in}\frac{[A_{out}]}{K_{in} + [A_{out}]}$<br/>
	<br/>
We are assuming that Lead transport through the transporter protein (pbrT) follows Michaelis-Menten kinetics. Lead influx through pbrT is also multiplied by $T_{mol}$, since each pbrT protein will aid in Lead influx. Number of Lead ions that can be sequestered by pbrD is not known hence we are denoting it by 'n'. In the calculation we have assumed n to be one.<br/>
	$\frac{d[A]}{dt} = T_{mol}v_{in}\frac{[A_{out}]}{K_{in} + [A_{out}]} + P([{A_{out}]-[A]}) - nk_{seq}[D][A]^n + nk_{-seq}[DA_n] - 2k_{dr1}[A]^2{[R_2]} + 2k_{-dr1}\frac{(T_A-[A_{out}]-[A]-n[DA_n])}{2} - 2k_{dr2}[A]^2(T_O - [O]) + 2k_{-dr2}[O]\frac{(T_A-[A_{out}]-[A]-n[DA_n])}{2}$ <br/>
	<br/>
Equations of $O$ and $F$ remain unchanged except for the resource competition incorporated in the equation of $F_{in}$.<br/>
	$\frac{d[O]}{dt} = - k_r{[R_2]}[O] + k_{-r}(T_O - [O]) -k_{dr2}[A]^2(T_O - [O]) + k_{-dr2}[O]\frac{(T_A-[A_{out}]-[A]-n[DA_n])}{2}$<br/><br/>
	$\frac{d[m_F]}{dt} = v_{TX1}\frac{[O]^2}{K_{TX1}^2 + [O]^2}-{\lambda_{m_F}}[m_F] + k_{leak}(T_O - [O])$
<br/><br/>
	$\frac{d[F_{in}]}{dt} = k_{TL1}[TLR]\frac{[m_F]}{{K_{TL1}}+{[m_F]}} - k_{mat}[F_{in}]$
<br/><br/>
	$\frac{d[F]}{dt} = k_{mat}[F_{in}] - kgd[F]$<br/><br/>
	
pbrD is also downstream of the operator (pbrAP), hence the free operator concentration can be used find level of pbrD expression. $\lambda_{m_D}$ is the first order decay rate of the pbrD protein. Leaky expression is same as before.<br/>
	$\frac{d[m_D]}{dt} = v_{TX3}\frac{[O]^2}{K_{TX3}^2 + [O]^2}-{\lambda_{m_D}[m_D]}+ k_{leak}(T_O - [O])$<br/>
	<br/>
	pbrD production is modelled using the resource competition model. $k_{seq}$ is the equilibrium constant for Lead sequestration and $k_{-seq}$ is the constant for the reverse reaction. <br/>
	$\frac{d[D]}{dt} = k_{TL3}[TLR]\frac{[m_D]}{{K_{TL3}}+{[m_D]}} - k_{seq}[D][A]^n + k_{-seq}[DA_n]$
<br/><br/>
	$\frac{d[DA_n]}{dt} = k_{seq}[D][A]^n - k_{-seq}[DA_n]$<br/><br/>
	pbrT is also downstream of the pbrR gene $(G_R)$ , hence the $G_R$ concentration can be used find level of pbrT expression. $\lambda_{m_T}$ is the first order decay rate of the pbrT protein.\newline
	$\frac{d[m_T]}{dt} = v_{TX4}\frac{[G_R]^2}{K_{TX4}^2 + [G_R]^2}-{\lambda_{m_T}[m_T]}$
<br/><br/>
	$\frac{d[T]}{dt} = k_{TL4}[TLR]\frac{[m_T]}{{K_{TL4}}+{[m_T]}}$
<br/><br/>
	Equation of TLR also remains unchanged as addition pbrD and pbrT has no effect on translation resources.
	$\frac{d[TLR]}{dt} = -v_{\lambda TLR}\frac{[TLR]}{{K_{\lambda TLR}}+[TLR]}$

## Lead Sequestration Time series

```python

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pylab as plot
from decimal import Decimal
import math as math
from scipy.integrate import solve_ivp

#Initialize some constants
vTX1 = 18.2       #reporter transcription rate constant
vTX2 = 18.2       #repressor transcription rate constant
KTX1 = 8.5        #Michaelis-Menten constant for reporter transcription 
KTX2 = 8.5        #Michaelis-Menten constant for repressor transcription
lam_m1 = 0.08     #reporter mRNA degradation rate constant
lam_m2 = 0.08     #repressor mRNA degradation rate constant
kTL1 = 0.082     #Reporter translation rate constant, 0.0076
kTL2 = 0.082     #Repressor translation rate constant, 0.0076
KTL1 = 29.9       #Michaelis-Menten constant for translation of reporter
KTL2 = 29.9       #Michaelis-Menten constant for translation of repressor
vlamTLR = 13.5    #Translation resources degradation rate constant
KlamTLR = 53.2    #Michaelis-Menten constant for degradation of TL resources
kmat = 0.1155      #reporter maturation rate constant
k2R = 50          #repressor dimerization rate constant
k_2R  = 0.001     #repressor dimer dissociation rate constant
kr = 960          #association rate constant for repression
k_r = 2.4         #dissociation rate constant for repression
kdr1 = 3.0e7      #association rate constant for first derepression mechanism 
k_dr1 = 12        #dissociation rate constant for first derepression mechanism
kdr2 = 3.0e7      #association rate constant for second derepression mechanism   
k_dr2 = 4800      #dissociation rate constant for second derepression mechanism
kleak = 0.0033    #leak reporter transcription rate constant
vTX3 = 8.2        #pbrD transcription rate constant
vTX4 = 8.2        #pbrT transcription rate constant
KTX3 = 8.5        #Michaelis-Menten constant for pbrD transcription
KTX4 = 8.5        #Michaelis-Menten constant for pbrT transcription
kTL3 = 0.082     #pbrD translation rate constant, 0.0076
kTL4 = 0.082     #pbrT translation rate constant, 0.0076
KTL3 = 29.9       #Michaelis-Menten constant for translation of pbrD
KTL4 = 1000       #Michaelis-Menten constant for translation of pbrT
lam_m3 = 0.08     #pbrD mRNA degradation rate constant
lam_m4 = 0.08     #pbrT mRNA degradation rate constant
P = 7.4*1e-4      #Permeability, for now using P of Na+ ~ 7.4*1e-4 nm/s
kgd = 0.001155

#DONOT KNOW VALUES OF CONSTANTS GIVEN BELOW----------\\\\\\\\\\|||||||||||||//////////----------\\\\\\\\\|||||||/////////
vin = 1e2         #Rate of Lead influx
Kin = 1e2         #Lead influx Michealis-Menten coefficient 
kseq = 1e3       #Lead Sequestration 
k_seq = 1e-15      #Reverse of 'kseq'

def model(z,t):
    
    #Initialize variables in a vector
    mT = z[0]
    T = z[1]
    Aout = z[2]
    A = z[3]          #analyte                       #R2A2 - repressor-operator complex
    mR = z[4]         #repressor mRNA
    R = z[5]          #repressor monomer
    R2 = z[6]         #repressor dimer               #R2O - free operator of reporter gene
    O = z[7]          #inactive repressor
    mF = z[8]         #reporter mRNA
    Fin = z[9]        #inactive reporter
    F = z[10]         #reporter
    mD = z[11]
    D = z[12]
    DA2 = z[13]
    TLR = z[14]       #Translation resources
    
    #ODEs  
    
    dmTdt = (vTX4*(GR**q)/((KTX4**q)+(GR**q))) - lam_m4*mT
    dTdt = (kTL4*TLR*(mT**2))/((KTL4**2)+(mT**2)) 
    
    T_mol = math.floor(T*0.6022)  #Volume of cell = 10^-15 L, Concentration [T] is in 10^-9 mol/L
    
    #LEAD TRANSPORT MODELLING ---------------------------->>>>>>>>>>>>>>><<<<<<<<<<<<<<<<--------------------------
    
    dAoutdt = -T_mol*(vin*(Aout-A))/(Kin + (Aout-A)) - P*(Aout - A)                                                                   #Lead Influx through pbrT
    dAdt = T_mol*(vin*(Aout-A))/(Kin + (Aout-A)) + P*(Aout-A) - n*kseq*(A**n)*D + n*k_seq*DA2 - 2*kdr1*(A**2)*R2 + 2*k_dr1*((TA-A-Aout-n*DA2)/2) - 2*kdr2*(A**2)*(TO-O) + 2*k_dr2*O*((TA-A-Aout-n*DA2)/2)  #Analyte Conc
    
    #LEAD TRANSPORT MODELLING ---------------------------->>>>>>>>>>>>>>><<<<<<<<<<<<<<<<--------------------------
    
    dmRdt = (vTX2*(GR**q)/((KTX2**q)+(GR**q))) - lam_m2*mR                                       #Repressor mRNA Conc
    dRdt = (kTL2*TLR*(mR))/(KTL2 + mR) - 2*k2R*(R**2) + 2*k_2R*R2             #Repressor Conc  
    dR2dt = k2R*(R**2) - k_2R*R2 - kr*R2*O + k_r*(TO-O) - kdr1*(A**2)*R2 + k_dr1*((TA-A-Aout-n*DA2)/2)      #Repressor dimer Conc               
    
    dOdt = -kr*R2*O + k_r*(TO-O) + kdr2*(A**2)*(TO-O) - k_dr2*O*((TA-A-Aout-n*DA2)/2)                       #Operator Conc
    dmFdt = (vTX1*(O**q)/((KTX1**q)+(O**q))) - lam_m1*mF + kleak*(TO-O)                          #Reporter mRNA Conc
    dFindt = (kTL1*TLR*(mF))/(KTL1 + mF) - kmat*Fin                           #Inactive Reporter Conc
    dFdt = kmat*Fin - kgd*F                                                                             #Active Reporter Conc  
    
    dmDdt = (vTX3*(O**q)/((KTX3**q)+(O**q))) - lam_m3*mD + kleak*(TO-O)
    dDdt = (kTL3*TLR*(mD**2))/((KTL3**2) + (mD**2))  - kseq*DA2
    dDA2dt = kseq*(A**n)*D - k_seq*DA2
    
    dTLRdt = -vlamTLR*TLR/((KlamTLR)+(TLR))                                                      #Translation Resources Conc
    
    return [dmTdt, dTdt, dAoutdt, dAdt, dmRdt, dRdt, dR2dt, dOdt, dmFdt, dFindt, dFdt, dmDdt, dDdt, dDA2dt, dTLRdt]


#Initial Conditions, conc in nM
mR0 = 0
R0 = 0
R20 = 0
R21 = 300
R2O0 = 0
R2A20 = 0
mF0 = 0
Fin0 = 0
F0 = 0
mD0 = 0
D0 = 0
DA20 = 0
mT0 = 0
T0 = 2
TLR0 = 1520

O0 = 5
GR = O0

Aout0 = 1e4
A0 = 0                 #Lead Concentration

n = 1       #Sequestration Number
p = 1       #Translation Hill function power
q = 2       #Transcription Hill function power

TA = Aout0 + A0 + 2*R2A20 + n*DA20
TO = O0 + R2O0        #LOOK AT THIS----------------------<<<<<<

t = np.linspace(0,250,100000)
z = [mT0,T0,Aout0,A0,mR0,R0,R20,O0,mF0,Fin0,F0,mD0,D0,DA20,TLR0]

ans = odeint(model, z, t)

LDout = ans[:,2]
LDin = ans[:,3]

plt.plot(t,LDout/1e3, 'r', label = 'Lead Outside')
plt.plot(t,LDin/1e3, 'g', label = 'Lead Inside')
plt.xlabel("Time (min)")
plt.ylabel("Lead Conc (nM)")
plt.legend()
plt.show()
```

```python
#TRIAL 2
#Change powers of the Hill function.

import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pylab as plot
from decimal import Decimal
import math as math
from scipy.integrate import solve_ivp
from matplotlib.pyplot import figure


#Initialize some constants
vTX1 = 18.2       #reporter transcription rate constant
vTX2 = 18.2       #repressor transcription rate constant
KTX1 = 8.5        #Michaelis-Menten constant for reporter transcription 
KTX2 = 8.5        #Michaelis-Menten constant for repressor transcription
lam_m1 = 0.08     #reporter mRNA degradation rate constant
lam_m2 = 0.08     #repressor mRNA degradation rate constant
kTL1 = 0.0082       #Reporter translation rate constant, 0.0076
kTL2 = 0.0082       #Repressor translation rate constant, 0.0076
KTL1 = 29.9       #Michaelis-Menten constant for translation of reporter
KTL2 = 29.9       #Michaelis-Menten constant for translation of repressor
vlamTLR = 13.5    #Translation resources degradation rate constant
KlamTLR = 53.2    #Michaelis-Menten constant for degradation of TL resources
kmat = 0.1155      #reporter maturation rate constant
kgd = 0.001155
k2R = 50          #repressor dimerization rate constant
k_2R  = 0.001     #repressor dimer dissociation rate constant
kr = 960          #association rate constant for repression
k_r = 2.4         #dissociation rate constant for repression
kdr1 = 3.0e7      #association rate constant for first derepression mechanism 
k_dr1 = 12        #dissociation rate constant for first derepression mechanism
kdr2 = 3.0e7      #association rate constant for second derepression mechanism   
k_dr2 = 4800      #dissociation rate constant for second derepression mechanism
kleak = 0.0033    #leak reporter transcription rate constant
vTX3 = 8.2        #pbrD transcription rate constant
vTX4 = 8.2        #pbrT transcription rate constant
KTX3 = 8.5        #Michaelis-Menten constant for pbrD transcription
KTX4 = 8.5        #Michaelis-Menten constant for pbrT transcription
kTL3 = 0.082       #pbrD translation rate constant, 0.0076
kTL4 = 0.082       #pbrT translation rate constant, 0.0076
KTL3 = 29.9       #Michaelis-Menten constant for translation of pbrD
KTL4 = 29.9       #Michaelis-Menten constant for translation of pbrT
lam_m3 = 0.08     #pbrD mRNA degradation rate constant
lam_m4 = 0.08     #pbrT mRNA degradation rate constant
P = 7.4*1e-4      #Permeability, for now using P of Na+ ~ 7.4*1e-4 nm/s

#DONOT KNOW VALUES OF CONSTANTS GIVEN BELOW----------\\\\\\\\\\|||||||||||||//////////----------\\\\\\\\\|||||||/////////
vin = 1           #Rate of Lead influx
Kin = 1e2         #Lead influx Michealis-Menten coefficient 
kseq = 1e3        #Lead Sequestration 
k_seq = 1e-15     #Reverse of 'kseq'

def model(t,z):
    
    #Initialize variables in a vector
    mT = z[0]
    T = z[1]
    Aout = z[2]
    A = z[3]          #analyte                       #R2A2 - repressor-operator complex
    mR = z[4]         #repressor mRNA
    R = z[5]          #repressor monomer
    R2 = z[6]         #repressor dimer               #R2O - free operator of reporter gene
    O = z[7]          #inactive repressor
    mF = z[8]         #reporter mRNA
    Fin = z[9]        #inactive reporter
    F = z[10]         #reporter
    mD = z[11]
    D = z[12]
    DA2 = z[13]
    TLR = z[14]       #Translation resources
    kseq = z[15]
    
    #ODEs  
    
    dmTdt = (vTX4*(GR**q)/((KTX4**q)+(GR**q))) - lam_m4*mT
    dTdt = (kTL4*TLR*(mT**p))/((KTL3**p)+(mT**p)) 
    
    T_mol = math.floor(T*0.6022)  #Volume of cell = 10^-15 L, Concentration [T] is in 10^-9 mol/L
    
    #LEAD TRANSPORT MODELLING ---------------------------->>>>>>>>>>>>>>><<<<<<<<<<<<<<<<--------------------------
    
    dAoutdt = -T_mol*(vin*Aout)/(Kin + Aout) - P*(Aout - A)                                                                   #Lead Influx through pbrT
    dAdt = T_mol*(vin*Aout)/(Kin + Aout) + P*(Aout-A) - n*kseq*(A**n)*D + n*k_seq*DA2 - 2*kdr1*(A**2)*R2 + 2*k_dr1*((TA-A-Aout-n*DA2)/2) - 2*kdr2*(A**2)*(TO-O) + 2*k_dr2*O*((TA-A-Aout-n*DA2)/2)  #Analyte Conc
    
    #LEAD TRANSPORT MODELLING ---------------------------->>>>>>>>>>>>>>><<<<<<<<<<<<<<<<--------------------------
    
    dmRdt = (vTX2*(GR**q)/((KTX2**q)+(GR**q))) - lam_m2*mR                                       #Repressor mRNA Conc
    dRdt = (kTL2*TLR*(mR**p)/((KTL3**p)+(mR**p))) - 2*k2R*(R**2) + 2*k_2R*R2             #Repressor Conc  
    dR2dt = k2R*(R**2) - k_2R*R2 - kr*R2*O + k_r*(TO-O) - kdr1*(A**2)*R2 + k_dr1*((TA-A-Aout-n*DA2)/2)      #Repressor dimer Conc               
    
    dOdt = -kr*R2*O + k_r*(TO-O) + kdr2*(A**2)*(TO-O) - k_dr2*O*((TA-A-Aout-n*DA2)/2)                       #Operator Conc
    dmFdt = (vTX1*(O**q)/((KTX1**q)+(O**q))) - lam_m1*mF + kleak*(TO-O)                          #Reporter mRNA Conc
    dFindt = (kTL1*TLR*(mF**p)/((KTL3**p)+(mF**p))) - kmat*Fin                           #Inactive Reporter Conc
    dFdt = kmat*Fin - kgd*F                                                                             #Active Reporter Conc  
    
    dmDdt = (vTX3*(O**q)/((KTX3**q)+(O**q))) - lam_m3*mD + kleak*(TO-O)
    dDdt = (kTL3*TLR*(mD**p))/((KTL3**p)+(mD**p))  - kseq*DA2
    dDA2dt = kseq*(A**n)*D - k_seq*DA2
    
    dTLRdt = -vlamTLR*TLR/((KlamTLR)+(TLR))                                                      #Translation Resources Conc
    
    dKdt = 0
    
    return [dmTdt, dTdt, dAoutdt, dAdt, dmRdt, dRdt, dR2dt, dOdt, dmFdt, dFindt, dFdt, dmDdt, dDdt, dDA2dt, dTLRdt, dKdt]


#Initial Conditions, conc in nM
mR0 = 0
R0 = 0
R20 = 0
R21 = 300
R2O0 = 0
R2A20 = 0
mF0 = 0
Fin0 = 0
F0 = 0
mD0 = 0
D0 = 0
DA20 = 0
mT0 = 0
T0 = 2
TLR0 = 1520

O0 = 8
GR = O0

Aout0 = 30000
A0 = 0                 #Lead Concentration

n = 1       #Sequestration Number
p = 1       #Translation Hill function power
q = 2       #Transcription Hill function power

TA = Aout0 + A0 + 2*R2A20 + n*DA20
TO = O0 + R2O0        #LOOK AT THIS----------------------<<<<<<

Kseq = []
LD = []
LD1 = []
LD60 = []
LD601 = []

for i in range(0,50,5):
    
    k = 10**(-i)
    Kseq.append(k)
    z = [mT0,T0,Aout0,A0,mR0,R0,R20,O0,mF0,Fin0,F0,mD0,D0,DA20,TLR0,k]

    ans = solve_ivp(fun=model, t_span=(0,1000), y0=z, method = 'Radau', t_eval = np.linspace(0,1000,100000))

    LDout = ans['y'][2][5587]
    LDin = ans['y'][3][5587]
    d = (30000 - LDout - LDin)/1e3
    e = (30000 - LDout)/1e3
    LD.append(d)
    LD1.append(e)
    
    LDout60 = ans['y'][2][335]
    LDin60 = ans['y'][3][335]
    d60 = (30000 - LDout60 - LDin60)/1e3
    e60 = (30000 - LDout60)/1e3
    LD60.append(d60)
    LD601.append(e60)


        
figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.plot(Kseq,LD, 'go', label = 'Lead Sequestered in 1000min')
plt.plot(Kseq,LD60, 'bo', label = 'Lead Sequestered in 60min')
#plt.plot(Kseq,LD1, 'r', label = 'Total Lead inside the cell')
plt.xlabel("k_seq value")
plt.ylabel("Lead (μM)")
plt.title('Lead Sequestration in presence of 30μM Lead')
plt.legend()
plt.xscale('log')
plt.ylim(0,1.0)
plt.show()

```
